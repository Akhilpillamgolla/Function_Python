```python
def add(a, b):
    """
    this is a doc string.
    
    return sum of 2 numbers
    """
    return (a+b)  # it will return the value we can assign that to variable , print the value.
```

```python
# it is not printing, it's a return value
add(5, 3)
```

```python
# you can receive return value into a variable

c = add(5,5)

print(c)
```

```python
def product(lst):
    """
    lst : it is a list
    return : product of all elements in a list
    """
    s = 1
    for i in lst:
        s*=i
    return s
```

```python
res = product( [2,5,3,1,4] ) 
# if we dont provide any value then it will give error that 1 required positional arguments.

print(res)
```

## Function Arguments

   - **Arguments Vs. Parameters** 
   
    
    - default arguments
    - arbitary arguments (args)
    - keyword arguments (kwargs)
     

arguments : when you call the function and pass some value.

parameters : they are variable that are defined in the function definiton.

```python
def intro(name, nationality):   # parameters
    print("Hello", name)
    print("I'm", nationality)
```

```python
# without giving nationality it still works
intro('Akhil','Indian')  # argument
# if we dont give any of the parameter then we will get the type error. 
```

```python
# default nationality has been changed to US
intro("Pillamgolla", "US")
```

- **Arguments are further classified into :** 
    - default arguments
    - arbitary arguments (args)
    - keyword arguments (kwargs)
     

```python
# default parameter: 
def intro(name, nationality = 'India'):   # default parameters
    print("Hello", name)
    print("I'm from", nationality)
```

```python
intro('akhil') # we dont get type error
```

```python
# default nationality has been changed to US
intro("Pillamgolla", "Hyderabad")
```

**arbitary arguments :** it can receive, any number of arguments

```python
def random_func(*args): # when we write * and args that is arbitary arguments.
    print(args)
    print(type(args))
    for i in args:
        print(i**2)
```

```python
random_func(10, 8, 3, 2, 5, 1)  # this is stored in tuple format.
# values
```

**keyword arguments :** it also takes any any number of arguments and stores in form of dictionary.
parameter name is mandatory while calling function

```python
def introduction(**kwargs): # here we take 2 stars. 
    print(kwargs)
    print(type(kwargs))
    
    for k,v in kwargs.items():
        print(k, ":", v)
```

```python
introduction(name="Akhil", age=23, hobby=["camera", 'coding'], nationality = 'Indian')
# we can also have list of 2 things. 
```

```python
introduction(name="Pillamgolla", age=24, company ="Google")
```

**Map**

```python
def add_one(x):
    return x+1  # this will add 1 to given x
```

```python
add_one(100)
```

```python
print(lst)
```

```python
# approch 1 
new_lst = []

for i in lst:
    new_lst.append(i+1)   # adding one to each element
    
print(new_lst)
```

```python
# approch 2 - best approch 
map_ele = map(add_one, lst)  # (function name ,values that you want to pass)

print(map_ele) # map object 
```

```python
print(  list(map_ele) ) # map object  -> list 
# using list converter.
```

**Filter**

```python
print(lst)
```

```python
def is_even(x):
    return x%2 == 0  # if you change this to odd we will get odd number.
```

```python
is_even(5)
```

```python
filter_ele = filter(is_even, lst)   # function name , parameter 

print(filter_ele)  # filter object
```

```python
list(filter_ele)
```
